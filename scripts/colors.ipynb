```python
import compas
from compas.datastructures import Mesh
from compas_notebook.app import App
from compas.colors import Color
from compas.geometry import Translation
from random import random


viewer = App()
mesh = Mesh.from_obj(compas.get('faces.obj'))
viewer.add(mesh, facecolor=Color(1.0, 0.0, 0.0), linecolor=Color(0.0, 0.0, 1.0), pointcolor=Color(0.0, 1.0, 0.0), show_points=True, pointsize=0.1)

mesh2 = mesh.copy()
mesh2.transform(Translation.from_vector([11, 0, 0]))
facecolor = {face: Color.from_i(random()) for face in mesh.faces()}
linecolor = {edge: Color.from_i(random()) for edge in mesh.edges()}
pointcolor = {vertex: Color.from_i(random()) for vertex in mesh.vertices()}
viewer.add(mesh2, facecolor=facecolor, linecolor=linecolor, pointcolor=pointcolor, show_points=True, pointsize=0.1)

viewer.show()
```
